```python
# Part a: Information
'''
Author: Gia Duc Can - ID 101570606
Assignment 1 - COMP2152
'''
# Part b: Create 4 variables with data type comments
gym_member = "Alex Alliton"  # str
preferred_weight_kg = 20.5  # float
highest_reps = 25  # int
membership_active = True  # bool

# Part c: Create dictionary with workout stats
# Dictionary (dict) with string keys and tuple values
workout_stats = {
    "Alex": (30, 45, 20),  # (yoga, running, weightlifting)
    "Jamie": (40, 30, 50),
    "Taylor": (25, 35, 30)
}

# Part d: Calculate total workout minutes for each friend
# Create a list of friend names first to avoid changing dict during iteration
friend_names = list(workout_stats.keys())
for friend in friend_names:
    workouts = workout_stats[friend]
    total_minutes = sum(workouts)
    workout_stats[f"{friend}_Total"] = total_minutes

print("Updated workout_stats with totals:")
print(workout_stats)

# Part e: Create 2D list from dictionary
# List (2-dimensional nested list)
workout_list = []
for friend in friend_names:
    workouts = workout_stats[friend]
    workout_list.append(list(workouts))

print("\nWorkout list (2D):")
print(workout_list)

# Part f: Slice the workout_list
print("\nYoga and Running minutes for all friends:")
for i, row in enumerate(workout_list):
    print(f"Friend {i+1}: Yoga = {row[0]}, Running = {row[1]}")

print("\nWeightlifting minutes for last two friends:")
for i, row in enumerate(workout_list[-2:]):
    print(f"Friend {len(workout_list)-1+i}: Weightlifting = {row[2]}")

# Part g: Check if any friend has >= 120 minutes total
print("\nFriends with 120+ minutes:")
for friend in friend_names:
    total = workout_stats[f"{friend}_Total"]
    if total >= 120:
        print(f"Great job staying active, {friend}!")

# Part h: User input to search for a friend
friend_name = input("Enter a friend's name: ")

if friend_name in friend_names:
    workouts = workout_stats[friend_name]
    total = workout_stats[f"{friend_name}_Total"]
    print(f"\n{friend_name}'s workout minutes:")
    print(f"  Yoga: {workouts[0]} minutes")
    print(f"  Running: {workouts[1]} minutes")
    print(f"  Weightlifting: {workouts[2]} minutes")
    print(f"  Total: {total} minutes")
else:
    print(f"Friend {friend_name} not found in the records.")

# Part i: Find friend with highest and lowest total workout minutes
friends_totals = {}
for friend in friend_names:
    friends_totals[friend] = workout_stats[f"{friend}_Total"]

highest_friend = max(friends_totals, key=friends_totals.get)
lowest_friend = min(friends_totals, key=friends_totals.get)

print("\n=== Workout Summary ===")
print(f"Friend with highest total workout minutes: {highest_friend} ({friends_totals[highest_friend]} minutes)")
print(f"Friend with lowest total workout minutes: {lowest_friend} ({friends_totals[lowest_friend]} minutes)")
```
